Data Types, Variables, and Conversions
--------------------------------------

In programming, data types are a classification which tells the interpreter how the programmer intends to use the data. For example, the interpreter needs to know if the data the programmer entered is a number or a string. Although there are several different data types, we will focus only on the following: 

Integer - used to specify whole numbers (no decimals), such as 1, 2, 3, and so on.

If an integer is entered with a decimal, the interpreter ignores the decimal.

For example, 3.75 is interpreted as 3.

**Float** \- used to specify numbers that need a decimal value, such as 3.14159. 

**String** - any sequence of characters such as letters, numbers, symbols, or punctuation marks.

**Boolean** - any data type that has a value of either True or False.

Use the **type()** function to determine the data type
The four basic data types we will use are:

-       Integer

-       Float

-       String

-       Boolean

````python
type(98)
````

````python
type(98.6)
````

````python
type("Hi!")
````

````python
type(True)
````

Creating and Using a Variable
-----------------------------

The Boolean operator for determining whether two values are equal is the double equal sign (**\==**).

A single equal sign (**\=**) is used to assign a value to a variable. The variable can then be used in other commands to recall value.

-       A variable can then be called for other operations.

````python
x=3
x*5
````

````python
"Cisco" * x
````

Concatenate Multiple String Variables
-------------------------------------

Concatenation is the process of combining multiple strings into one string. For example, the concatenation of "**foot**" and "**ball**" is "**football**".

In the following example, four variables are concatenated together in a **print()** statement with the plus sign (**+**).

Notice that spaces are inserted between the strings 

````python
str1="Cisco"
str2="Networking"
str3="Academy"

print (str1,str2,str3)
````

By default, print() uses a separator of one space between arguments

Converting Data Types
---------------------

Concatenation does not work for different data types.

````python
x=3
print ("The value of X is " +x)
````

F-String was introduced in Python 3.6, and is now the preferred way of formatting strings.

- To specify a string as an f-string, simply put an f in front of the string literal, and add curly brackets {} as

  placeholders for variables and other operations.

````python
x=3
txt = f" This value of X is {x}"
print(txt)
````

**f-strings are generally preferred** in modern Python, especially for teaching, readability, and professional code.
An f-string shows exactly what the final output will look like:
easier to understand compared with using commas, concatenation, or format()
1. Clear and readable

An f-string shows exactly what the final output will look like:

````python
print(f"{str1} {str2} {str3}")
````

### **2\. Faster and more efficient**

Python evaluates f-strings at runtime and they are typically **faster** than older formatting methods.
### **3\. No accidental spaces**

Using commas: Adds automatic spaces, which is sometimes unwanted.

````python
print(str1, str2, str3)
````

Using f-strings gives full control:

````python
print(f"{str1}-{str2}-{str3}")
````

### **4\. Works well with expressions**

You can put variables and expressions directly inside:

````python
print(f"{str1.upper()} {len(str2)}")
````

````python
print(f"{str1[0]}"
      f"{str1[1]}"
      f"{str1[2:]} "
      f"{str2[:-1]}"
      f"{str2[-1]} "
      f"{str3.capitalize()[:-1] + str3[-1].upper()}")
````

##### Formatting  Data Types with python
```
-You can also format a value directly without keeping it in a variable:
-Use "{:.2f}" to display a float to two decimal places.
-Change the 2 to increase or decrease decimal places.

```

````python
pi= 22/7
print(pi)
````

````python

print(f"{pi:.2f}")
````
